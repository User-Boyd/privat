```python
import pandas as pd
import requests
import json
```

```python
response = requests.get("https://api-server.dataquest.io/economic_data/countries?filter_by=region=Sub-Saharan%20Africa")
wb_data = response.json()
```

```python
print(wb_data)
```

```python
data = json.loads(wb_data)
df = pd.DataFrame(data)
df.columns
```

```python
cols = [ 'country_code', 'short_name', 'region', '2-alpha_code', 
         'currency_unit', 'income_group', 'latest_population_census', 'latest_household_survey', 
         'latest_agricultural_census', 'latest_industrial_data'
       ]
df = df[cols] 
df
```
